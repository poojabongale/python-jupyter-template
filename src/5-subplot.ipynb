### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (top-left)
axs[0, 0].plot(x, y, marker='o', color='blue', linestyle='-')
axs[0, 0].set_title("Line Chart")

# Bar chart (top-right)
axs[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'purple'])
axs[0, 1].set_title("Bar Chart")

# Scatter plot (bottom-left)
axs[1, 0].scatter(x, y, color='blue')
axs[1, 0].set_title("Scatter Plot")

# Pie chart (bottom-right)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'purple'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```
